# TASK-4- Image to Pencil Sketch with Python:

We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.


# Importing Libraries

```python
import cv2 #OpenCV-Python is designed to solve computer vision problems.
import matplotlib.pyplot as plt
```

# Loading and Plotting the original image

```python
img = cv2.imread("flower.jpeg")

#Converting image from BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize = (7,7))
plt.imshow(img) #for plotting
plt.title("Original Image")
plt.show()
```

# Converting original image to grayscale

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.figure(figsize = (7,7))
plt.imshow(img_gray, cmap="gray")
plt.title("GrayScale Image")
plt.show()
```

# Inverting the image

```python
img_invert = cv2.bitwise_not(img_gray) #helps in studying the image in more detail
plt.figure(figsize = (7,7))
plt.imshow(img_invert, cmap = "gray") #for plotting
plt.title("Inverted Image")
plt.axis("off")
plt.show()
```

# Smoothing the image

```python
#smoothing reduces the noise and produce less pixelated images 
img_smoothing = cv2.GaussianBlur(img_invert, (23,23), sigmaX=0, sigmaY = 0)
plt.figure(figsize= (7,7))
plt.imshow(img_smoothing, cmap="gray")
plt.title("Blur Image") #smooth and less edgy image
plt.axis("off")
plt.show()
```

# Converting image to pencil sketch

```python
#Grayscale image has pixel values 0-255 
#use previous output to convert into pencil sketch
final_img = cv2.divide(img_gray, 255-img_smoothing, scale = 255)
plt.figure(figsize = (7,7))
plt.imshow(final_img, cmap="gray")
plt.title("Final Sketch Image")
plt.axis("off")
plt.show()
```

# Final Output

```python
#using subplot to disply all the images in one frame

plt.figure(figsize = (18,18))
plt.subplot(1,5,1)
plt.title("Original Image")
plt.axis("off")
plt.imshow(img)


plt.subplot(1,5,2)
plt.imshow(img_gray, cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")


plt.subplot(1,5,3)
plt.imshow(img_invert, cmap = "gray")
plt.axis("off")
plt.title("Inverted Image")

plt.subplot(1,5,4)
plt.imshow(img_smoothing, cmap = "gray")
plt.axis("off")
plt.title("Blur Image")

plt.subplot(1,5,5)
plt.imshow(final_img, cmap = "gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()

```
